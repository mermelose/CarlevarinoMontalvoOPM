## **Descripción de la Base de Datos**

Estos datos abordan el logro de los estudiantes en la educación secundaria de dos escuelas portuguesas.

Los atributos de los datos incluyen calificaciones de los estudiantes, características demográficas, sociales y relacionadas con la escuela, los cuales fueron recopilados empleando informes escolares y cuestionarios.

Este conjunto de datos es sobre el rendimiento de los alumnos en la materia de Matemáticas.

**Información de los atributos:**

- *school*: escuela del estudiante (binario: 'GP' - Gabriel Pereira o 'MS' - Mousinho da Silveira)
- *sex*: sexo del estudiante (binario: 'F' - femenino o 'M' - masculino)
- *age*: edad del estudiante (numérico: de 15 a 22)
- *address*: tipo de dirección del hogar del estudiante (binario: 'U' - urbano o 'R' - rural)
- *famsize*: tamaño de la familia (binario: 'LE3' - menor o igual a 3 o 'GT3' - mayor que 3)
- *Pstatus*: estado de convivencia de los padres (binario: 'T' - viviendo juntos o 'A' - separados)
- *Medu*: educación de la madre (numérico: 0 - ninguna, 1 - educación primaria (4° grado), 2 - de 5° a 9° grado, 3 - educación secundaria o 4 - educación superior)
- *Fedu*: educación del padre (numérico: 0 - ninguna, 1 - educación primaria (4° grado), 2 - de 5° a 9° grado, 3 - educación secundaria o 4 - educación superior)
- *Mjob*: trabajo de la madre (nominal: 'teacher' - profesora, 'health' - relacionado con la salud, 'services' civiles (ej. administrativo o policía), 'at_home' - en casa o 'other' - otro)
- *Fjob*: trabajo del padre (nominal: 'teacher' - profesor, 'health' - relacionado con la salud, 'services' civiles (ej. administrativo o policía), 'at_home' - en casa o 'other' - otro)
- *reason*: motivo para elegir esta escuela (nominal: cerca de 'home' - casa, reputación de la escuela 'reputation', preferencia de 'course' - curso u 'other' - otro)
- *guardian*: tutor del estudiante (nominal: 'mother' - madre, 'father' - padre u 'other' - otro)
- *traveltime*: tiempo de viaje de casa a la escuela (numérico: 1 - <15 min., 2 - 15 a 30 min., 3 - 30 min. a 1 hora, o 4 - >1 hora)
- *studytime*: tiempo de estudio semanal (numérico: 1 - <2 horas, 2 - 2 a 5 horas, 3 - 5 a 10 horas, o 4 - >10 horas)
- *failures*: número de fracasos académicos previos (numérico: n si 1<=n<3, de lo contrario 4)
- *schoolsup*: apoyo educativo adicional (binario: sí o no)
- *famsup*: apoyo educativo familiar (binario: sí o no)
- *paid*: clases pagadas adicionales dentro de la materia (Matemáticas o Portugués) (binario: sí o no)
- *activities*: actividades extracurriculares (binario: sí o no)
- *nursery*: asistió a la escuela infantil (binario: sí o no)
- *higher*: quiere cursar estudios superiores (binario: sí o no)
- *internet*: acceso a Internet en casa (binario: sí o no)
- *romantic*: tiene una relación romántica (binario: sí o no)
- *famrel*: calidad de las relaciones familiares (numérico: de 1 - muy mala a 5 - excelente)
- *freetime*: tiempo libre después de la escuela (numérico: de 1 - muy bajo a 5 - muy alto)
- *goout*: salir con amigos (numérico: de 1 - muy bajo a 5 - muy alto)
- *Dalc*: consumo de alcohol entre semana (numérico: de 1 - muy bajo a 5 - muy alto)
- *Walc*: consumo de alcohol en fin de semana (numérico: de 1 - muy bajo a 5 - muy alto)
- *health*: estado de salud actual (numérico: de 1 - muy malo a 5 - muy bueno)
- *absences*: número de ausencias escolares (numérico: de 0 a 93)

```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

# Definir el modelo
modelAdam = Sequential()

# Añadir la primera capa densa (input layer)
modelAdam.add(Dense(units=64, activation='relu', input_dim=X_train.shape[1]))

# Añadir una segunda capa densa (hidden layer)
modelAdam.add(Dense(units=32, activation='relu'))

# Capa de salida con activación sigmoid
modelAdam.add(Dense(units=1, activation='sigmoid'))  # Sigmoid para salida entre 0 y 1

# Compilar el modelo con el optimizador Adam
modelAdam.compile(optimizer=Adam(), loss=keras.losses.mean_squared_error)

# Mostrar el resumen del modelo
modelAdam.summary()
```

```python
# Entrenar el modelo
historyAdam = modelAdam.fit(X_train, Y_train / 20, epochs=100, batch_size=32, validation_data=(X_val, Y_val / 20),random_state = 32)

# Evaluar el modelo en el conjunto de prueba
test_loss_Adam = modelAdam.evaluate(X_test, Y_test / 20)
print(f'Test loss: {test_loss_Adam**(1/2)*20}')
```

```python
import matplotlib.pyplot as plt

# Graficar la pérdida de entrenamiento y validación
plt.plot(historyAdam.history['loss'], label='Entrenamiento')
plt.plot(historyAdam.history['val_loss'], label='Validación')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.legend()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
try:
    import scipy.optimize as spo
except Exception as e:
    spo = None


class BFGSMLPTrainer:
    """
    Minimal trainer that builds a Keras MLP and fits its weights using SciPy's BFGS / L-BFGS-B.
    Usage:
        trainer = BFGSMLPTrainer(in_dim=X.shape[1], hidden=[64,32])
        res = trainer.fit(X_train, y_train, X_val=X_val, y_val=y_val, method='L-BFGS-B', options={'maxiter':1000})
        preds = trainer.predict(X_test)
    The trainer stores per-evaluation training loss and (optionally) validation loss so you can plot them.
    """
    def __init__(self, in_dim, hidden=[64], activation='relu', output_activation='sigmoid', loss_fn=None, dtype=tf.float32):
        self.in_dim = in_dim
        self.hidden = list(hidden)
        self.activation = activation
        self.output_activation = output_activation
        self.dtype = dtype
        self._cache_w = None
        self._cache_loss = None
        self._cache_grad = None
        # counter for SciPy function evaluations (useful for debugging/printing)
        self._eval_count = 0
        # whether to print per-eval diagnostics (default: False)
        self.print_evals = False
        # store numeric losses per SciPy evaluation so user can plot or inspect
        self.eval_history_train = []
        self.eval_history_val = []
        # optional validation data (set in fit)
        self._X_val = None
        self._y_val = None
        if loss_fn is None:
            self.loss_fn = tf.keras.losses.MeanSquaredError()
        else:
            self.loss_fn = loss_fn
        self._build_model()

    def _build_model(self):
        from keras import Sequential
        from keras.layers import Dense, Input
        model = Sequential()
        model.add(Input(shape=(self.in_dim,)))
        for h in self.hidden:
            model.add(Dense(h, activation=self.activation, dtype=self.dtype))
        model.add(Dense(1, activation=self.output_activation, dtype=self.dtype))
        # build to initialize weights
        model.build((None, self.in_dim))
        self.model = model

    def get_weights_vector(self):
        weights = self.model.get_weights()
        self._shapes = [w.shape for w in weights]
        vec = np.concatenate([w.ravel() for w in weights]).astype(np.float64)
        return vec

    def set_weights_vector(self, vec):
        new_weights = []
        idx = 0
        for s in self._shapes:
            size = int(np.prod(s))
            arr = vec[idx:idx+size].reshape(s).astype(np.float32)
            new_weights.append(arr)
            idx += size
        self.model.set_weights(new_weights)

    def _compute_loss_and_grad(self, vec, X, y):
        # set weights first
        self.set_weights_vector(vec)
        X_tf = tf.convert_to_tensor(np.asarray(X).astype(np.float32))
        y_tf = tf.convert_to_tensor(np.asarray(y).astype(np.float32))
        with tf.GradientTape() as tape:
            preds = self.model(X_tf, training=False)
            # compute MSE and then take sqrt to obtain RMSE
            mse = tf.reduce_mean(keras.losses.mean_squared_error(y_tf, preds))
            loss = tf.sqrt(mse)
        # convert loss (RMSE) to python float for caching and plotting
        loss_value = float(loss.numpy())
        # increment evaluation counter
        self._eval_count += 1
        # append to training history (store RMSE)
        self.eval_history_train.append(loss_value)
        # compute validation loss if validation data was provided
        val_loss_value = None
        if self._X_val is not None and self._y_val is not None:
            try:
                Xv_tf = tf.convert_to_tensor(np.asarray(self._X_val).astype(np.float32))
                yv_tf = tf.convert_to_tensor(np.asarray(self._y_val).astype(np.float32))
                preds_val = self.model(Xv_tf, training=False)
                val_mse = tf.reduce_mean(keras.losses.mean_squared_error(yv_tf, preds_val))
                val_loss = tf.sqrt(val_mse)
                val_loss_value = float(val_loss.numpy())
                # store validation RMSE
                self.eval_history_val.append(val_loss_value)
            except Exception:
                # if validation eval fails, append None to keep indices aligned
                self.eval_history_val.append(None)
                val_loss_value = None
        # optionally print a short diagnostic
        if self.print_evals:
            try:
                # print RMSE diagnostics
                if val_loss_value is None:
                    print(f"SciPy-eval {self._eval_count}: train_RMSE={loss_value:.6f}")
                else:
                    print(f"SciPy-eval {self._eval_count}: train_RMSE={loss_value:.6f}, val_RMSE={val_loss_value:.6f}")
            except Exception:
                pass
        grads = tape.gradient(loss, self.model.trainable_variables)
        if grads is None:
            grad_vec = np.zeros_like(self.get_weights_vector())
        else:
            grad_vec = np.concatenate([g.numpy().ravel().astype(np.float64) for g in grads])
        # cache
        self._cache_w = np.array(vec, copy=True)
        # cache numeric loss
        self._cache_loss = float(loss_value)
        self._cache_grad = grad_vec
        return self._cache_loss, self._cache_grad

    def _fun_loss(self, w, X, y):
        # SciPy may call fun and jac separately; use cache to avoid double compute
        if self._cache_w is not None and np.array_equal(self._cache_w, w):
            return float(self._cache_loss)
        loss, grad = self._compute_loss_and_grad(w, X, y)
        return float(loss)

    def _fun_grad(self, w, X, y):
        if self._cache_w is not None and np.array_equal(self._cache_w, w):
            return self._cache_grad
        loss, grad = self._compute_loss_and_grad(w, X, y)
        return grad

    def fit(self, X, y, X_val=None, y_val=None, method='BFGS', options={'maxiter':100, 'disp': True}, verbose=True, print_evals=False):
        """
        Fit the MLP weights using SciPy's optimizer. X and y should be array-like.
        If X_val and y_val are provided, the trainer will compute validation loss at each SciPy evaluation and store it.

        Parameters:
            X_val, y_val: optional validation data (array-like). If provided, per-eval val loss will be computed and stored.
            print_evals: bool - if True, print per-evaluation diagnostics (default False)
        Returns the SciPy OptimizeResult object.
        """
        X_np = np.asarray(X)
        y_np = np.asarray(y)
        if y_np.ndim == 1:
            y_np = y_np.reshape(-1, 1)
        # store validation data (kept as numpy arrays)
        if X_val is not None and y_val is not None:
            self._X_val = np.asarray(X_val)
            self._y_val = np.asarray(y_val)
            if self._y_val.ndim == 1:
                self._y_val = self._y_val.reshape(-1, 1)
        else:
            self._X_val = None
            self._y_val = None
        w0 = self.get_weights_vector()
        # clear cache
        self._cache_w = None
        # reset eval counter and set printing flag from call
        self._eval_count = 0
        self.print_evals = bool(print_evals)
        # reset eval histories
        self.eval_history_train = []
        self.eval_history_val = []
        fun = lambda w: self._fun_loss(w, X_np, y_np)
        jac = lambda w: self._fun_grad(w, X_np, y_np)
        res = spo.minimize(fun, w0, jac=jac, method=method, options=options)
        # set final weights
        self.set_weights_vector(res.x)
        if verbose:
            print('Optimization success:', res.success, 'message:', res.message)
        return res

    def predict(self, X):
        X_np = np.asarray(X).astype(np.float32)
        preds = self.model.predict(X_np)
        return preds.flatten()

    def get_eval_history(self):
        """Return a dict with 'train' and 'val' lists (val may be empty if no validation provided)."""
        return {'train': list(self.eval_history_train), 'val': list(self.eval_history_val)}
```

```python
# Preparar arrays numpy para train/val/test
X_train_np = np.asarray(X_train).astype(np.float32)
y_train_np = np.asarray(Y_train / 20.0).astype(np.float32).reshape(-1, 1)
X_val_np = np.asarray(X_val).astype(np.float32)
y_val_np = np.asarray(Y_val / 20.0).astype(np.float32).reshape(-1, 1)
X_test_np = np.asarray(X_test).astype(np.float32)
y_test_np = np.asarray(Y_test / 20.0).astype(np.float32).reshape(-1, 1)
```

```python
trainer = BFGSMLPTrainer(in_dim=X_train_np.shape[1], hidden=[64, 32])
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
opts = {'maxiter': 100, 'disp': False}

res = trainer.fit(X_train_np, y_train_np, X_val=X_val_np, y_val=y_val_np, method='BFGS', options=opts, print_evals=True)
```

```python
preds = trainer.predict(X_test_np)
rmse_scaled = root_mean_squared_error(y_test_np, preds)
print(f"BFGS-B test RMSE : {rmse_scaled*20:.4f}")
```
